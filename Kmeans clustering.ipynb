```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler

"""
Created on Tue Sep 20 13:42:11 2022

@author: Junhyun
"""

def EDA_kmeans(X, k):
    """

    Parameters
    ----------
    X : array
        데이터
    k : int
        클러스터의 개수

    Returns
    -------
    clusterX.

    """
    if isinstance(X, np.ndarray):
        X = pd.DataFrame(X)
    
    # Standard Scaler
    scaler = StandardScaler()
    scaler.fit(X)
    scaled_X = scaler.transform(X)
    
    kmeans = KMeans(n_clusters=k, random_state=0).fit(scaled_X)
    clusterX = np.concatenate([X, kmeans.labels_.reshape(len(X),1)], axis=1)
    
    colname = X.columns.tolist()
    colname.append('Cluster')
    clusterX = pd.DataFrame(clusterX, columns=colname)
    
    return(clusterX)

# =============================================================================
# from sklearn.datasets import load_iris
# X = load_iris(as_frame=True)['data']
# kdat = EDA_kmeans(X, k=3)
# =============================================================================
```

```python
from sklearn.datasets import load_iris
X = load_iris(as_frame=True)['data']
col = [X.columns, 'Cluster']
```

```python
kdat = EDA_kmeans(X, k=3)
```
