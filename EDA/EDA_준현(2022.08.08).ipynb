```python
# 계량시간, 보압 절환 압력, 사출최대속도, 사출최대압력, 쿠션위치, 히터4, 히터6
```

```python
dd = df.iloc[22342:,:]
```

```python
outIdx = np.where((dd['불량이름']==out1[1]))[0]   
```

```python
import matplotlib as mpl
plt.rc('font', family='NanumGothic')
mpl.rcParams['axes.unicode_minus'] = False

col = ['blue', 'black', 'red','darkorange', 'green']
plt.figure(figsize=(20,10))
plt.scatter(np.arange(len(dd)), dd['사출최대속도'].values, alpha=0.5, label='사출최대속도')
plt.scatter(np.arange(len(dd)), dd['사출최대압력'].values, alpha=0.5, label='사출최대압력')
for i in range(len(out1)):
    outIdx = np.where((dd['불량이름']==out1[i]))[0]    
    plt.axvline(x=outIdx[0], color=col[i], linestyle='--', linewidth=2, label=out1[i])
    for j in range(1,len(outIdx)):
        plt.axvline(x=outIdx[j], color=col[i], linestyle='--', linewidth=2)
plt.legend()

plt.show()
```
